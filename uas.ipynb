```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv("pangan.csv")

df['Tahun'] = df['Tahun'].astype(int)
df['Bulan'] = df['Bulan'].astype(int)
df['Harga'] = df['Harga'].astype(float)

df_minyak = df[df['Komoditas'].str.contains("Minyak Goreng Kemasa", case=False, na=False)].copy()

df_minyak['time'] = df_minyak['Tahun'] + (df_minyak['Bulan'] - 1) / 12


X = df_minyak[['time']]
y = df_minyak['Harga']

model = LinearRegression()
model.fit(X, y)

print("Intercept:", model.intercept_)
print("Koefisien:", model.coef_[0])

print(f"Harga = {model.intercept_:.2f} + {model.coef_[0]:.2f} * time")

next_time = df_minyak['time'].max() + 1/12
prediksi = model.predict([[next_time]])
print("Prediksi harga bulan depan:", prediksi[0])
```
