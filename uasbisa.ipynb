```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv("pangann.csv")

df["Harga"] = df["Harga"].replace("-", None)
df = df.dropna(subset=["Harga"])
df["Harga"] = df["Harga"].str.replace("Rp", "").str.replace(",", ".").astype(float)

df_minyak = df[df["Komoditas"].str.contains("Minyak Goreng Kemasan", case=False, na=False)].copy()

if df_minyak.empty:
    print("Data Minyak Goreng Kemasan tidak ada.")
else:

    X = df_minyak[["Tahun"]]
    y = df_minyak["Harga"]

    model = LinearRegression()
    model.fit(X, y)

    print("Persamaan regresi:")
    print(f"Harga = {model.intercept_:.2f} + {model.coef_[0]:.2f} * Tahun")

    next_year = df_minyak["Tahun"].max() + 1
    prediksi = model.predict([[next_year]])[0]

    prediksi_rp = f"Rp{prediksi:,.0f}".replace(",", ".")
    print(f"\nPrediksi harga tahun depan: {prediksi_rp}")
```
